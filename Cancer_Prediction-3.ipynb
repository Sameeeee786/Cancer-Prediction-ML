# **Cancer Prediction**

Dataset Information:

Target Variable (y):
- Diagnosis (M = malignant, B = benign)

Ten features (X) are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter)
2. texture (standard deviation of gray-scale values)
3. perimeter
4. area
5. smoothness (local variation in radius lengths)
6.  compactness (perimeter^2 / area - 1.0)
7. concavity (severity of concave portions of the contour)
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension (coastline approximation - 1)

For each characteristic three measures are given:

a. Mean

b. Standard error

c. Largest/ Worst

## ðŸ“Š Project Overview

This project is about detecting whether a tumor is malignant or benign using a machine learning model.  
We are using the Breast Cancer Wisconsin dataset and applying **Logistic Regression** for classification.

The model predicts:
- ðŸŸ¢ Benign (non-cancerous tumor)
- ðŸ”´ Malignant (cancerous tumor)

Such predictions can help in early diagnosis and better treatment planning.

```python
# Step 1 : import library
import pandas as pd
```

```python
# Step 2 : import data
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
#
cancer.describe()

#
print("\nðŸ§¹ Missing Value Check:\n")
print(cancer.isnull().sum())
```

```python
# Step 3 : define target (y) and features (X)
```

```python
cancer.columns
```

```python
y = cancer['diagnosis']
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
